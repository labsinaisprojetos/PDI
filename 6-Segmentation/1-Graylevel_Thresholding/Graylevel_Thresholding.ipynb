Gray-level thresholding

Loading modules  and image

```python
import cv2
import matplotlib.pyplot as plt
#the input 0 means that the image is being read on grayscale
img = cv2.imread('peter.png',0);
```

On the code above we are thresholding the image and using its threshold as an mask

```python
ret,thresh=cv2.threshold(img,105,255,cv2.THRESH_BINARY_INV)
img2=cv2.bitwise_and(img,img,mask=thresh)
cv2.imwrite('graylevel_threshold.jpg',thresh)
cv2.imwrite('img_masked.png',img2)
```

```python
plt.imshow(img, cmap='gray')
plt.title('Original image')
plt.axis('off')
plt.show()
plt.imshow(thresh, cmap='gray')
plt.title('Threshold')
plt.axis('off')
plt.show()
plt.imshow(img2, cmap='gray')
plt.title('Using threshold as an mask on original image')
plt.axis('off')
plt.show()
```
