Example: chain link fence hole detection



Loading modules and image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Loading image, converting color to grayscale and using threshold with otsu's method

```python
img=cv2.imread("fence.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,img_thresh=cv2.threshold(img_gray,125,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

Creating an (151,151) cross matrix as kernel and eroding the image

```python
kernel=np.zeros((151,151),np.uint8)
kernel[76,:]=1
kernel[:,76]=1
img_eroded=cv2.erode(img_thresh,kernel,iterations=1)
cv2.imwrite('thresholded_image.jpg',img_thresh)
cv2.imwrite('eroded_image.jpg',img_eroded)
```

Plotting the images

```python
plt.imshow(img,cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
plt.imshow(img_thresh,cmap='gray')
plt.title("Thresholded image")
plt.axis('off')
plt.show()
plt.imshow(img_eroded,cmap='gray')
plt.title('Erosion with an 151x151 cross matrix')
plt.axis('off')
plt.show()
```
