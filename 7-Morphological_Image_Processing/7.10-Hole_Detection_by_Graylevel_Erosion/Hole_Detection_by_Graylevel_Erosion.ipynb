Example: chain link fence hole detection



Loading modules and image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("fence.jpg")
```

Eroding the image on gray level and thresholding the eroded image

```python
img=cv2.imread("fence.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
kernel=cv2.getStructuringElement(cv2.MORPH_CROSS,(151,151))
img_eroded=cv2.erode(img_gray,kernel,iterations=1)
ret,img_thresh=cv2.threshold(img_eroded,175,255,cv2.THRESH_BINARY)
cv2.imwrite('eroded.jpg',img_eroded)
cv2.imwrite('threshold_of_the_eroded_image.jpg',img_thresh)
```

Plotting the image

```python
plt.imshow(img,cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
plt.imshow(img_eroded,cmap='gray')
plt.title("Erosion with a 151x151 cross matrix as kernel")
plt.axis('off')
plt.show()
plt.imshow(img_thresh,cmap='gray')
plt.title('Threshold of the eroded image')
plt.axis('off')
plt.show()
```
