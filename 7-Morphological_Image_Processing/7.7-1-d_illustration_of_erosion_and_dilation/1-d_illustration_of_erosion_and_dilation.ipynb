1-d illustration of erosion and dilation




Loading modules and image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('peter.png')
```

Plotting 1d-graphs of the original signal,dilated signal and eroded signal with variable kernels lenghts

```python
original_signal=img[180, 150:249]
kernel=np.ones((1,1),np.uint8)
a=np.ones((1,2),np.uint8)
for i in range(1,23,2):
    dilated_signal = cv2.dilate(img,kernel,iterations=1)
    dilated_signal=dilated_signal[180,150:249]
    eroded_signal=cv2.erode(img,kernel,iterations=1)
    eroded_signal=eroded_signal[180,150:249]
    plt.plot(original_signal, 'b--')
    plt.plot(dilated_signal, 'r')
    plt.plot(eroded_signal, 'y')
    plt.title('Kernel length ={}'.format(str(i)))
    plt.savefig('1D_erosion_dilation_{}.png'.format(str(i)))
    plt.show()
    kernel=np.hstack([kernel,a])
```
