Loading modules and image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Loading image and getting its edges with morphological operations

```python
img=cv2.imread("cliparts.png")
img=cv2.resize(img,(500,360))
# 3x3 elipse element
kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
img_eroded=cv2.erode(img,kernel,iterations=1)
img_dilated=cv2.dilate(img,kernel,iterations=1)
edge1 = img_dilated - img;
edge2 = img - img_eroded;
edge3 = edge1 + edge2;
cv2.imwrite('edge_dilated.jpg',edge1)
cv2.imwrite('edge_eroded.jpg',edge2)
cv2.imwrite('edge_dilated_and_eroted.jpg',edge3)
```

Plotting the images

```python
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
plt.figure(figsize=(7,5))
plt.imshow(edge1)
plt.title("Edge1= dilated - original")
plt.axis('off')
plt.show()
plt.figure(figsize=(7,5))
plt.imshow(edge2)
plt.title('Edge2= eroded - original')
plt.axis('off')
plt.show()
plt.figure(figsize=(7,5))
plt.imshow(edge3)
plt.title('Edge3= edge1+ edge2')
plt.axis('off')
plt.show()
```
