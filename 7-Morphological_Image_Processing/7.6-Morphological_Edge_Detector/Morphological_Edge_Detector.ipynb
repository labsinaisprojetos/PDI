Loading modules and image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

First we will load the image, then we will get a 3x3 ellipse-shapped structure and use it as kernel to make a eroded version of the image. Then we will make an dilated version of the image. We will proceed to subtract the original image from the dilated one and call it "edge1", then we will do the same with the eroded one and call it "edge2". Be summing "edge1" and "edges2" we get an image that shows the edges with a highter accuracy to the image original shape

```python
img=cv2.imread("cliparts.png")
img=cv2.resize(img,(500,360))
# 3x3 elipse element
kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
img_eroded=cv2.erode(img,kernel,iterations=1)
img_dilated=cv2.dilate(img,kernel,iterations=1)
edge1 = img_dilated - img;
edge2 = img - img_eroded;
edge3 = edge1 + edge2;
cv2.imwrite('edge_dilated.jpg',edge1)
cv2.imwrite('edge_eroded.jpg',edge2)
cv2.imwrite('edge_dilated_and_eroted.jpg',edge3)
```

Plotting the images

```python
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
plt.figure(figsize=(9,6.5))
plt.imshow(edge1)
plt.title("Edge1= dilated - original")
plt.axis('off')
plt.show()
plt.figure(figsize=(9,6.5))
plt.imshow(edge2)
plt.title('Edge2= eroded - original')
plt.axis('off')
plt.show()
plt.figure(figsize=(9,6.5))
plt.imshow(edge3)
plt.title('Edge3= edge1+ edge2')
plt.axis('off')
plt.show()
```
