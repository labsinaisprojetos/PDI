Cascaded dilations



Loading modules and image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("butterfly.png")
```

We will be defining 3 different kernels and then proceed to perform 3 multiple sequences of dilation operations on the image

```python
kernel1=np.ones((1,10),np.uint8)
kernel2=np.identity(7,np.uint8)
kernel3=np.fliplr(kernel2)
img_dilated = cv2.dilate(img,kernel1,iterations=1)
img_dilated2 = cv2.dilate(img_dilated,kernel2,iterations=1)
img_dilated3 = cv2.dilate(img_dilated2,kernel3,iterations=3)
cv2.imwrite('Cascaded_Dilations_1.jpg',img_dilated)
cv2.imwrite('Cascaded_Dilations_2.jpg',img_dilated2)
cv2.imwrite('Cascaded_Dilations_3.jpg',img_dilated3)
```

Plotting images

```python
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(img_dilated)
plt.title("")
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(img_dilated2)
plt.title("")
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(img_dilated3)
plt.title("")
plt.axis('off')
plt.show()
```
