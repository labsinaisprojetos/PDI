Loading modules

```python
import csv
import matplotlib.pyplot as plt
```

Reading csv file

```python
with open('linCIE2008v2e_1.csv') as csvfile:
    readfile=csv.reader(csvfile,quoting=csv.QUOTE_NONNUMERIC)
    wavelenght=[]
    efficiency=[]
    peak=[0.0,0.0]
    for row in readfile:
        if row[1]>peak[1]:
            peak=row
        wavelenght.append(row[0])
        efficiency.append(row[1])
```

Plotting graph

```python
plt.xlabel('Wavelenght (nm)')
plt.ylabel('Efficiency')
plt.title('10-deg function, stepsize : 5nm')
plt.plot(wavelenght,efficiency,'-b',label='Luminous function',)
plt.plot(peak[0],peak[1],'^r',label='Peak value')
plt.axvline(x=peak[0], color='black', linestyle=':',linewidth=0.5)
plt.axhline(y=peak[1], color='black', linestyle=':',linewidth=0.5)
plt.text(peak[0],peak[1],'Peak value('+str(peak[0])+','+str(peak[1])+')',Fontsize='12')
plt.legend(loc='center right')
```
