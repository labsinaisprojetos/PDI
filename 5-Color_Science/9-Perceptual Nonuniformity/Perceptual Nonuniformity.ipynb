```python
x = data[:, 0] * 1000
y = data[:, 1] * 1000
a = data[:, 2] *10
b = data[:, 3] *10
t = data[:, 4] / 180 * np.pi
f = np.arange(0, 2 * np.pi, 0.0001)
```

```python
for i in range(data.shape[0]):
    aa=a[i] * np.cos(f)
    bb=b[i] * np.sin(f)
    xx=x[i] + aa*np.cos(t[i])-bb*np.sin(t[i])
    yy=y[i] + aa*np.sin(t[i])+bb*np.cos(t[i])
    ax.plot(xx, yy, 'k', linewidth=1);
    ax.plot(xx+2, yy+2, 'w', linewidth=1);
```
