```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
%matplotlib inline
```

First we get the data

```python
mat = loadmat('coneFundamentals.mat')
coneFundamentals = mat['coneFundamentals']
coneFundamentals.ndim
```

Get the information that we want

```python
wavelength = coneFundamentals[:,0]
LMS = coneFundamentals[:, 1:4]
```

Plotting the data

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel(r'Wavelength $\lambda$ (nm)', fontsize=15)
ax.set_ylabel('Normalized Sensitivity', fontsize=15)
ax.set_title('Cone Fundamentals', fontsize=15)
red, = ax.plot(wavelength, LMS[:,0], color='red', label='red')
green, = ax.plot(wavelength, LMS[:,1], color='green', label='green')
blue, = ax.plot(wavelength, LMS[:,2], color='blue', label='blue')
ax.legend([red, green, blue], [r'$L: S_{R}({\lambda})$', r'$M: S_{G}({\lambda})$', r'$S: S_{B}({\lambda})$'])
ax.text(445, 0.9, '445 nm', fontsize=15, color='blue');
ax.text(535, 0.9, '535 nm', fontsize=15, color='green');
ax.text(575, 0.95, '575 nm', fontsize=15, color='red');
ax.axis([390, 760, 0, 1])
ax.grid()
```
