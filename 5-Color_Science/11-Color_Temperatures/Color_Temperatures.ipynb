```python
from PIL import Image
from IPython.display import display

img = Image.open('chromadiagram.png')
temp_table = {
    1000: (255, 56, 0),
    2000: (255, 137, 18),
    3000: (255, 180, 107),
    4000: (255, 209, 163),
    5000: (255, 228, 206),
    6000: (255, 243, 239),
    7000: (245, 243, 255),
    8000: (227, 233, 255),
    9000: (214, 225, 255),
    11000: (196, 215, 255),
    13000: (186, 208, 255),
    15000: (179, 204, 255),
    20000: (168, 197, 255),
    25000: (163, 193, 255)}
tempK = [1000,3000,5000,7000,9000,11000,13000,15000,20000,25000]

for i in range(0, len(tempK)):
    r,g,b = temp_table[tempK[i]]
    color_matrix = (r / 255.0, 0.0, 0.0, 0.0, 0.0, g / 255.0, 0.0, 0.0, 0.0, 0.0, b / 255.0, 0.0)
    img2 = img.convert('RGB', color_matrix)
    img2.save('img_'+str(tempK[i])+'K_.jpg', "JPEG", quality=80, optimize=True, progressive=False)
    print(str(tempK[i]))
    display(img2)
```
