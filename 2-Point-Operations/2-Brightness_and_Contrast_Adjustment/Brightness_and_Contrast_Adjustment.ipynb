Loading modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Loading image

```python
img = cv2.imread('parrot.jpg')
img = cv2.normalize(img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
```

Adjusting image brightness by intensity scaling

```python
scale = 1.2
scaledImg = scale*img

for i in range(scaledImg.shape[0]):
    for j in range(scaledImg.shape[1]):
        for h in range(scaledImg.shape[2]):
            if scaledImg[i][j][h] > 1:
                scaledImg[i][j][h] = 1
```

Plot the original image and the intensity scaled image

```python
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1, 3, 2)
plt.imshow(scaledImg)
plt.title('Scaled Image')
cv2.imwrite('Brightness_scaled.png', scaledImg*255)
```

Adjusting image contrast by changing its gamma value

```python
gamma = 1.5
gammaImg = img**gamma
```

Normalizying the image with new gamma value

```python
gammaImg *= 255.0/gammaImg.max()
```

Plotting the image with new gamma

```python
plt.subplot(1, 3, 3)
plt.imshow(gammaImg.astype(np.uint8))
plt.title('Gamma Image')
cv2.imwrite('Contrast_gamma.png', gammaImg)

plt.show()
```
