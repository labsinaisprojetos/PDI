```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
%matplotlib inline
```

First we get the data

```python
mat = loadmat('ciexyz31_1.mat')
cmfs = mat['CMFs']
cmfs.ndim
```

Get the information that we want

```python
wavelength = cmfs[:,0]
xyz = cmfs[:, 1:4]
```

Plotting the data

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel(r'Wavelength $\lambda$ (nm)', fontsize=15)
ax.set_ylabel('Tristimulus values', fontsize=15)
ax.set_title('XYZ color matching functions', fontsize=15)
red, = ax.plot(wavelength, xyz[:,0], color='red', label='red')
green, = ax.plot(wavelength, xyz[:,1], color='green', label='green')
blue, = ax.plot(wavelength, xyz[:,2], color='blue', label='blue')
ax.legend([red, green, blue], [r'$x({\lambda})$', r'$y({\lambda})$', r'$z({\lambda})$'])
ax.axis([300, 800, 0, 2])
ax.grid()
```

Finnaly we save the figure

```python
fig.savefig('XYZ_Matching_Functions_xyzmatching.png')
```
