## Adaptive Histogram Equalization

Implemented in Python3 by Roberto Pontes


```python
# Imports
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = '/home/roberto/codes/PDI/4-Histograms/' \
       '4-Adaptive-Histogram-Equalization/'
```

```python
# Load image and convert to gray scale
image_name = 'parrot.jpg'  # 'dental.jpg' 'skull.jpg'
image = cv.imread(PATH + image_name)

image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

### Image Equalization


```python
# Global
image_eq = cv.equalizeHist(image)
```

```python
# Clahe 8x8
clahe = cv.createCLAHE(tileGridSize=(8, 8))
image_clahe8_8 = clahe.apply(image)
```

```python
# Clahe 16x16
clahe = cv.createCLAHE(tileGridSize=(16, 16))
image_clahe16_16 = clahe.apply(image)
```

```python
# For plot images
f, axarr = plt.subplots(2, 2)
axarr[0, 0].axis("off")
axarr[0, 0].set_title(' Image(s)')
axarr[0, 0].imshow(cv.cvtColor(image, cv.COLOR_GRAY2RGB))
axarr[0, 1].axis("off")
axarr[0, 1].set_title(' Image(s)')
axarr[0, 1].imshow(cv.cvtColor(image_eq, cv.COLOR_GRAY2RGB))
axarr[1, 0].axis("off")
axarr[1, 0].set_title(' Image(s)')
axarr[1, 0].imshow(cv.cvtColor(image_clahe8_8, 
                               cv.COLOR_GRAY2RGB))
axarr[1, 1].axis("off")
axarr[1, 1].set_title(' Image(s)')
axarr[1, 1].imshow(cv.cvtColor(image_clahe16_16, 
                               cv.COLOR_GRAY2RGB))
```

```python
res = np.hstack((image,image_eq))
res2 = np.hstack((image_clahe8_8, image_clahe16_16))
res = np.vstack((res, res2))

plt.imshow(res)
```
