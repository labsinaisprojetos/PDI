```python
# Imports
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = '/home/roberto/codes/PDI/4-Histograms/1-Generating-Histogram/'
```

```python
# Load image
image = cv.imread(PATH + 'bay.jpg')
# image = cv.imread(PATH + 'brain.jpg')
```

```python

# Plot image
plt.figure(1)
plt.axis("off")
plt.imshow(image)
```

```python
# Calculate histogram
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
hist, bins = np.histogram(image_gray.ravel(), 256, [0, 256])
```

```python
# Plots
plt.bar(range(0, 256), hist)
plt.title('Histogram')
plt.xlim([0, 256])
plt.xlabel('Gray level')
plt.ylabel('#pixel')
```
