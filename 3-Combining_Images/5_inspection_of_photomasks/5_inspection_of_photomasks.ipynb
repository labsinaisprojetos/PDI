```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

Supose we got two aligned images from two circuits, in which one is correct and the other has printing issues.

```python
maskImg1 = cv2.imread("mask1.png",0)
maskImg2 = cv2.imread("mask2.png",0)
figs, axs = plt.subplots(1, 2)
titlelist = ['original circuit','circuit with defect']
imagelist = [maskImg1,maskImg2]
for ax,title,fig in zip(axs,titlelist,imagelist):
    ax.imshow(fig,vmax=255,vmin=0)
    ax.set_title(title.capitalize())
    ax.axis('off')
```

It is hard to see differences between images by looking at both, but simply subtracting one mask from the other will render an new masking containing the deffects

```python
diffImg = abs(maskImg1 - maskImg2)
plt.gray()
plt.imshow(diffImg);
plt.axis('Off')
plt.title('Difference Image')
cv2.imwrite('Mask_Comparison_diff.png',diffImg)
```

Now it is easier to do photomasks inspection. But **be aware**, if masks are not aligned this won't work
